## Transform

### Dealing with 'indicator' and 'country' columns

From above, it was observed that the 'indicator' and 'country' columns has dictionary as values. A function will be created to extract the 'values' from those dictionaries.

```python
#creating a function to extract values
def extract_columns(df):
    indicator_list = []
    country_list = []
    for i in range(len(df)):
        indicator_list.append(df.iloc[i,0]['value'])

    for c in range(len(df)):
        country_list.append(df.iloc[c,1]['value'])

    #drop and replace indicator and country columns

    df_out = df.drop(['indicator', 'country'], axis=1)

    df_out.insert(0,column='indicator', value=indicator_list)
    df_out.insert(1,column='country', value=country_list)

    return df_out
```

#### Creating a unified dataframe

```python
#function
def merge_dataframes_init(df_1, df_2):
    #drop columns
    df_1 = df_1.drop(['unit', 'obs_status', 'decimal'], axis=1)
    df_2 = df_2.drop(['countryiso3code', 'unit', 'obs_status', 'decimal'], axis=1)

    #get indicator type
    df_1_indi = df_1['indicator'].unique()[0]
    df_2_indi = df_2['indicator'].unique()[0]

    #merge
    df_merged = pd.merge(left=df_1, right=df_2, how='left', on=['country', 'date'])

    #rename 'value_x' and 'value_y' columns
    df_merged = df_merged.rename(columns={'value_x':df_1_indi, 'value_y':df_2_indi})
    
    #drop 'indicator_x' and 'indicator_y' columns
    df_merged = df_merged.drop(['indicator_x', 'indicator_y'], axis=1)
    

    return df_merged
```

```python
#creating the unified dataframe(population and gdp)
df_econ_indi = merge_dataframes_init(df_population, df_gdp)
df_econ_indi
```

```python
#creating functions to merge df_econ_indi to other dataframes
def merge_all(df):
    #transfering df_econ_indi to local variable
    df_base = df_econ_indi

    #drop columns of input dataframe
    df = df.drop(['countryiso3code', 'unit', 'obs_status', 'decimal'], axis=1)

    #get df indicator name
    df_indi = df['indicator'].unique()[0]

    #merge dataframes
    df_base = df_base.merge(df, how='left', on=['country', 'date'])

    #rename 'value' column
    df_base = df_base.rename(columns={'value':df_indi})

    #dropping 'indicator' columns
    df_base = df_base.drop(['indicator'], axis=1)

    return df_base
```

```python
#merge with df_gdp_growth
df_econ_indi = merge_all(df_gdp_gwt)
df_econ_indi
```

```python
#merge with df_cpi
df_econ_indi = merge_all(df_cpi)
df_econ_indi
```

```python
#merge with df_unemployment
df_econ_indi = merge_all(df_unemployment)
df_econ_indi
```
